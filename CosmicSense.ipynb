```python
import pandas as pd
```

```python
dataset=pd.read_csv('/content/star_classification.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.head(5)
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
features=['alpha','delta','u','g','r','i','z','redshift']
dataset[features]=scaler.fit_transform(dataset[features])
```

```python
X=dataset[features]
Y=dataset['class']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
#logistic regression

from sklearn.linear_model import LogisticRegression

model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import classification_report,accuracy_score

print(classification_report(Y_test,y_pred))
```

```python
print(f'Accuracy {accuracy_score(Y_test,y_pred)}')
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={'n_estimators':[50,100,200],'max_depth':[None,10,20,30]}
```

```python
from sklearn.ensemble import RandomForestClassifier

grid_search=GridSearchCV(estimator=RandomForestClassifier(),param_grid=param_grid,cv=3,scoring='accuracy',n_jobs=-1)
```

```python
grid_search.fit(X_train,Y_train)
```

```python
print(f"best Parameters:{grid_search.best_params_}")
print(f"best Parameters:{grid_search.best_score_}")
```

```python
y_pred=model.predict(X_test)
print(f'AccuracyScore : {accuracy_score(Y_test,y_pred)}')
```
